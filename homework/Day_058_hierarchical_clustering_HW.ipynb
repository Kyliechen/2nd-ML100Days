```python
# 載入套件
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
```

```python
# 設定 2D 樣板資料
n_samples = 1500
random_state = 100

# 生成 雙半月 資料點
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)

# 生成 緊密三群 資料點
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)

# 生成 2維均勻分布 資料點
no_structure = np.random.rand(n_samples, 2), None
```

```python
# 設定群集與繪圖參數
plt.figure(figsize=(10, 8))
default_base = {'n_neighbors': 10, 'n_clusters': 3}
datasets = [(noisy_moons, {'n_clusters': 2}), (blobs, {'n_clusters': 3}), (no_structure, {'n_clusters': 3})]
```

# 作業
* 試著完成剩下三種 2D 樣板資料的繪圖, 並觀察有何不同

```python
# 執行各種樣板資料的繪圖迴圈
plot_num = 1
for idx, (data, param) in enumerate(datasets):
    X, y = data
    X = StandardScaler().fit_transform(X)
    
    ward = cluster.AgglomerativeClustering(n_clusters=param['n_clusters'], linkage="ward")
    comp = cluster.AgglomerativeClustering(n_clusters=param['n_clusters'], linkage="complete")
    aveg = cluster.AgglomerativeClustering(n_clusters=param['n_clusters'], linkage="average")
    sing = cluster.AgglomerativeClustering(n_clusters=param['n_clusters'], linkage="single") 

    clustering_algorithms = (('Average Linkage', aveg),
                             ('Complete Linkage', comp),
                             ('Ward Linkage', ward),
                             ('Single Linkage', sing))
    
    for name, algo in clustering_algorithms:
        algo.fit(X)
        
        if hasattr(algo, 'labels_'):
            y_pred = algo.labels_.astype(np.int)
        else:
            y_pred = algo.predict(X)
        
        plt.subplot(len(datasets), len(clustering_algorithms), plot_num)

        if idx == 0:
            plt.title(name, size=10)
        
        colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00', '#4daf4a',
                                             '#f781bf', '#a65628', '#984ea3',
                                             '#999999', '#e41a1c', '#dede00']),
                                      int(max(y_pred) + 1))))
        plt.scatter(X[:, 0], X[:, 1], s=10, color=colors[y_pred])

        plt.xlim(-2.5, 2.5)
        plt.ylim(-2.5, 2.5)
        plt.xticks(())
        plt.yticks(())
        plot_num += 1
plt.show()
```

Sinlge linkage has the best performance in noisy_moons set. However, it has the worst performance in no_structure set. For blobs, all the linkages look good.
